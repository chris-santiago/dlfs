```python
import numpy as np
import dlfs.base as dlfs
import matplotlib.pyplot as plt
```

```python
x = np.arange(-3, 3, 0.01)
```

```python
chain_1 = [dlfs.square, dlfs.sigmoid]
y = dlfs.eval_two_chain(chain_1, x)
dy_dx = dlfs.deriv_two_chain(chain_1, x)
plt.plot(x, y)
plt.plot(x, dy_dx)
plt.legend(["f(x)", "$\\frac{dy}{dx}$"])
plt.show()
```

```python
chain_2 = [dlfs.leaky_relu, dlfs.sigmoid, dlfs.square]
y = dlfs.eval_three_chain(chain_2, x)
dy_dx = dlfs.deriv_three_chain(chain_2, x)
plt.plot(x, y)
plt.plot(x, dy_dx)
plt.legend(["f(x)", "$\\frac{dy}{dx}$"])
plt.show()
```
